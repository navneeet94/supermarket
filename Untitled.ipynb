# PLOT SALES BY CITY

```python
city_obj = df.groupby(by="City")
```

```python
city_df = city_obj.first()
```

```python
city_obj.get_group('York')
```

```python
city_names = city_df.index.tolist()
city_names
```

```python
city_dict = {"city":[],"sales":[]}
for i in city_names:
    print(i)    
    city_dict["city"].append(i)
   
    # sales sum
    unique_city_sales = city_obj.get_group(i)['Sales'].values.sum()
    print(unique_city_sales)
    city_dict["sales"].append(unique_city_sales)
    
    print('---------------------------------------')
```

```python
city_sales_df = pd.DataFrame(city_dict)
city_sales_df
```

```python
city_sales_df[(city_sales_df['sales'] >= 30000)]
```

```python
plt.figure(figsize=(800,50))
sns.barplot(x=city_sales_df['city'],y=city_sales_df['sales']);
```

# PLOT WITH ATLEAST GREATER THAN THAT NUMBER CITY SALES

```python
at_least_val = int(input("Enter number for greater than this number :- "))
city_at_least_df = city_sales_df[(city_sales_df['sales'] >= at_least_val)]
plt.figure(figsize=(20,5))
sns.barplot(x=city_at_least_df['city'],y=city_at_least_df['sales']);
```
